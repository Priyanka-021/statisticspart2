  The ANOVA test (Analysis of Variance) is a statistical method used to determine whether there are statistically significant differences between the means of three or more independent groups.   Assumptions of ANOVA

  2.Normality- The data in each group are approximately normally distributed.

  3.Homogeneity of variances (homoscedasticity)- All groups have equal
     variances.Can be tested using Levene's Test or Bartlett's Tes

**3.Write a Python function to calculate the probability distribution function (PDF) of a Bernoulli distribution.**


```python
def bernoulli_pdf(x, p):
    if x not in (0, 1):
        raise ValueError("x must be 0 or 1 for a Bernoulli distribution.")
    if not 0 <= p <= 1:
        raise ValueError("p must be a probability in the range [0, 1].")

    return p if x == 1 else 1 - p
print(bernoulli_pdf(1, 0.7))

```

**9.Write a Python function to calculate Z-scores from a dataset and plot them.**

```python
import numpy as np
import matplotlib.pyplot as plt

def calculate_and_plot_z_scores(data):

    data = np.array(data)
    mean = np.mean(data)
    std = np.std(data)

    if std == 0:
        raise ValueError("Standard deviation is zero. Z-scores are undefined.")

    z_scores = (data - mean) / std

    plt.figure(figsize=(10, 5))
    plt.plot(z_scores, marker='o', linestyle='-', color='blue')
    plt.axhline(0, color='black', linestyle='--', linewidth=1)
    plt.axhline(1, color='green', linestyle='--', linewidth=1)
    plt.axhline(-1, color='green', linestyle='--', linewidth=1)
    plt.axhline(2, color='orange', linestyle='--', linewidth=1)
    plt.axhline(-2, color='orange', linestyle='--', linewidth=1)
    plt.axhline(3, color='red', linestyle='--', linewidth=1)
    plt.axhline(-3, color='red', linestyle='--', linewidth=1)
    plt.title('Z-scores of Dataset')
    plt.xlabel('Index')
    plt.ylabel('Z-score')
    plt.grid(True)
    plt.show()


data = np.random.normal(100, 15, 50)
z_scores = calculate_and_plot_z_scores(data)
```

**14.Write a Python program to calculate the Z-score for a given data point and compare it to a standard normal distribution.**


```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

def calculate_and_plot_z_score(x, mean, std_dev):

    z = (x - mean) / std_dev
    print(f"Z-score for x = {x} (mean = {mean}, std = {std_dev}): {z:.3f}")

    p = norm.cdf(z)
    print(f"Probability of observing a value ‚â§ {x}: {p:.4f} ({p*100:.2f}%)")


    x_vals = np.linspace(-4, 4, 1000)
    y_vals = norm.pdf(x_vals)

    plt.figure(figsize=(10, 5))
    plt.plot(x_vals, y_vals, color='blue', label='Standard Normal PDF')
    plt.title('Standard Normal Distribution')
    plt.xlabel('Z-score')
    plt.ylabel('Probability Density')

    x_fill = np.linspace(-4, z, 1000)
    y_fill = norm.pdf(x_fill)
    plt.fill_between(x_fill, y_fill, color='skyblue', alpha=0.5, label=f'P(Z ‚â§ {z:.2f})')

    plt.axvline(z, color='red', linestyle='--', label=f'Z = {z:.2f}')
    plt.legend()
    plt.grid(True)
    plt.show()
```

**15.Implement hypothesis testing using Z-statistics for a sample dataset.**


```python
import math
from scipy.stats import norm

mu_0 = 70
sigma = 10
x_bar = 74
n = 30
alpha = 0.05


z = (x_bar - mu_0) / (sigma / math.sqrt(n))


z_critical = norm.ppf(1 - alpha)

p_value = 1 - norm.cdf(z)

if z > z_critical:
    result = "Reject the null hypothesis: There is evidence that the new method improves scores."
else:
    result = "Fail to reject the null hypothesis: No sufficient evidence for improvement."

print(f"Z-statistic: {z:.3f}")
print(f"Z-critical (0.05 level): {z_critical:.3f}")
print(f"P-value: {p_value:.4f}")
print(result)
```

**16.Create a confidence interval for a dataset using Python and interpret the result.**


```python
import numpy as np
import scipy.stats as stats

data = [72, 75, 78, 70, 74, 77, 69, 73, 71, 76]


mean = np.mean(data)
sem = stats.sem(data)

confidence = 0.95
n = len(data)
df = n - 1
t_critical = stats.t.ppf((1 + confidence) / 2, df)

margin_of_error = t_critical * sem

lower_bound = mean - margin_of_error
upper_bound = mean + margin_of_error

print(f"Sample Mean = {mean:.2f}")
print(f"95% Confidence Interval = ({lower_bound:.2f}, {upper_bound:.2f})")
```

**1.Write a Python program to perform a Z-test for comparing a sample mean to a known population mean and interpret the results.**


```python
import numpy as np
from scipy.stats import norm

def z_test(sample_data, population_mean, population_std, alpha=0.05):

    n = len(sample_data)
    sample_mean = np.mean(sample_data)


    z_stat = (sample_mean - population_mean) / (population_std / np.sqrt(n))

    p_value = 2 * (1 - norm.cdf(abs(z_stat)))


    print(f"Sample mean = {sample_mean:.4f}")
    print(f"Population mean = {population_mean}")
    print(f"Z-statistic = {z_stat:.4f}")
    print(f"P-value = {p_value:.4f}")


    if p_value < alpha:
        print(f"Reject the null hypothesis at Œ± = {alpha}.")
        print("There is significant evidence that the sample mean differs from the population mean.")
    else:
        print(f"Fail to reject the null hypothesis at Œ± = {alpha}.")
        print("There is not enough evidence to say the sample mean differs from the population mean.")

    return z_stat, p_value


np.random.seed(0)
sample = np.random.normal(loc=52, scale=10, size=40)
pop_mean = 50
pop_std = 10

z_test(sample, pop_mean, pop_std, alpha=0.05)
```

**4.Perform a two-tailed Z-test using Python and visualize the decision region on a plot.**


```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

def two_tailed_z_test_with_plot(sample, pop_mean, pop_std, alpha=0.05):
    n = len(sample)
    sample_mean = np.mean(sample)

    z_stat = (sample_mean - pop_mean) / (pop_std / np.sqrt(n))


    p_value = 2 * (1 - norm.cdf(abs(z_stat)))

    print(f"Sample mean = {sample_mean:.4f}")
    print(f"Population mean = {pop_mean}")
    print(f"Z-statistic = {z_stat:.4f}")
    print(f"P-value = {p_value:.4f}")

    z_crit = norm.ppf(1 - alpha / 2)
    print(f"Critical Z-value for alpha={alpha}: ¬±{z_crit:.4f}")

    x = np.linspace(-4, 4, 1000)
    y = norm.pdf(x)

    plt.figure(figsize=(10, 6))
    plt.plot(x, y, label='Standard Normal Distribution', color='blue')

    plt.fill_between(x, 0, y, where=(x <= -z_crit), color='red', alpha=0.5, label='Rejection Region (Left)')
    plt.fill_between(x, 0, y, where=(x >= z_crit), color='red', alpha=0.5, label='Rejection Region (Right)')

    plt.axvline(z_stat, color='green', linestyle='--', linewidth=2, label=f'Z Statistic = {z_stat:.2f}')

    plt.title('Two-tailed Z-test Decision Regions')
    plt.xlabel('Z-value')
    plt.ylabel('Probability Density')
    plt.legend()
    plt.grid(True)
    plt.show()

    if abs(z_stat) > z_crit:
        print("Reject the null hypothesis: Sample mean significantly differs from population mean.")
    else:
        print("Fail to reject the null hypothesis: No significant difference detected.")

    return z_stat, p_value
np.random.seed(1)
sample_data = np.random.normal(loc=102, scale=15, size=40)  # Sample data
population_mean = 100
population_std = 15

two_tailed_z_test_with_plot(sample_data, population_mean, population_std, alpha=0.05)
```

**5.Create a Python function that calculates and visualizes Type 1 and Type 2 errors during hypothesis testing.**


```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

def plot_type1_type2_errors(mu0, mua, sigma, n, alpha=0.05):

    se = sigma / np.sqrt(n)


    z_crit = norm.ppf(1 - alpha/2)


    crit_low = mu0 - z_crit * se
    crit_high = mu0 + z_crit * se

    x = np.linspace(mu0 - 4*se, mua + 4*se, 1000)

    pdf_H0 = norm.pdf(x, loc=mu0, scale=se)
    pdf_Ha = norm.pdf(x, loc=mua, scale=se)

    plt.figure(figsize=(12,6))


    plt.plot(x, pdf_H0, label=r'$H_0$: $\mu = %.2f$' % mu0, color='blue')
    plt.plot(x, pdf_Ha, label=r'$H_a$: $\mu = %.2f$' % mua, color='orange')


    plt.fill_between(x, 0, pdf_H0, where=(x <= crit_low), color='red', alpha=0.3, label='Type 1 Error Region (Œ±)')
    plt.fill_between(x, 0, pdf_H0, where=(x >= crit_high), color='red', alpha=0.3)


    plt.fill_between(x, 0, pdf_Ha, where=( (x > crit_low) & (x < crit_high) ), color='green', alpha=0.3, label='Type 2 Error Region (Œ≤)')


    plt.axvline(crit_low, color='red', linestyle='--')
    plt.axvline(crit_high, color='red', linestyle='--')

    plt.title('Type 1 (Œ±) and Type 2 (Œ≤) Errors in Hypothesis Testing')
    plt.xlabel('Sample Mean')
    plt.ylabel('Probability Density')
    plt.legend()
    plt.grid(True)
    plt.show()


    beta = norm.cdf(crit_high, loc=mua, scale=se) - norm.cdf(crit_low, loc=mua, scale=se)

    print(f"Significance level (Type 1 error Œ±): {alpha:.3f}")
    print(f"Type 2 error (Œ≤): {beta:.3f}")
    print(f"Power of the test (1 - Œ≤): {1 - beta:.3f}")

    return alpha, beta, 1 - beta

plot_type1_type2_errors(mu0=100, mua=105, sigma=15, n=36, alpha=0.05)
```

**6.Write a Python program to perform an independent T-test and interpret the results.**


```python
import numpy as np
from scipy.stats import ttest_ind

def independent_t_test(sample1, sample2, alpha=0.05):

    t_stat, p_value = ttest_ind(sample1, sample2, equal_var=True)

    print(f"Group 1 mean = {np.mean(sample1):.3f}, std = {np.std(sample1, ddof=1):.3f}, n = {len(sample1)}")
    print(f"Group 2 mean = {np.mean(sample2):.3f}, std = {np.std(sample2, ddof=1):.3f}, n = {len(sample2)}")
    print(f"T-statistic = {t_stat:.4f}")
    print(f"P-value = {p_value:.4f}")

    if p_value < alpha:
        print(f"Reject the null hypothesis at Œ± = {alpha}.")
        print("There is a statistically significant difference between the two groups.")
    else:
        print(f"Fail to reject the null hypothesis at Œ± = {alpha}.")
        print("There is no statistically significant difference between the two groups.")

    return t_stat, p_value

np.random.seed(123)
group1 = np.random.normal(loc=20, scale=5, size=30)
group2 = np.random.normal(loc=22, scale=5, size=30)

independent_t_test(group1, group2, alpha=0.05)
```

**7.Perform a paired sample T-test using Python and visualize the comparison results.**


```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import ttest_rel

def paired_t_test_with_plot(before, after, alpha=0.05):

    t_stat, p_value = ttest_rel(before, after)

    print(f"Mean before: {np.mean(before):.3f}")
    print(f"Mean after: {np.mean(after):.3f}")
    print(f"T-statistic: {t_stat:.4f}")
    print(f"P-value: {p_value:.4f}")

    if p_value < alpha:
        print(f"Reject the null hypothesis at Œ± = {alpha}.")
        print("Significant difference between before and after measurements.")
    else:
        print(f"Fail to reject the null hypothesis at Œ± = {alpha}.")
        print("No significant difference detected.")


    plt.figure(figsize=(8,5))


    for i in range(len(before)):
        plt.plot([1, 2], [before[i], after[i]], marker='o', color='gray', alpha=0.6)


    means = [np.mean(before), np.mean(after)]
    std_errors = [np.std(before, ddof=1)/np.sqrt(len(before)),
                  np.std(after, ddof=1)/np.sqrt(len(after))]

    plt.errorbar([1, 2], means, yerr=std_errors, fmt='o', color='red', capsize=5, label='Mean ¬± SE')

    plt.xticks([1, 2], ['Before', 'After'])
    plt.ylabel('Measurements')
    plt.title('Paired Sample Comparison')
    plt.legend()
    plt.grid(True)
    plt.show()

    return t_stat, p_value

np.random.seed(42)
before_treatment = np.random.normal(loc=50, scale=5, size=20)
after_treatment = before_treatment + np.random.normal(loc=-3, scale=4, size=20)

paired_t_test_with_plot(before_treatment, after_treatment)
```

**9.Write a Python function to calculate the confidence interval for a sample mean and explain its significance.**


```python
import numpy as np
from scipy.stats import t

def confidence_interval(sample, confidence=0.95):

    n = len(sample)
    mean = np.mean(sample)
    std_err = np.std(sample, ddof=1) / np.sqrt(n)


    t_crit = t.ppf((1 + confidence) / 2, df=n-1)

    margin_of_error = t_crit * std_err
    lower_bound = mean - margin_of_error
    upper_bound = mean + margin_of_error

    return mean, lower_bound, upper_bound


np.random.seed(123)
sample_data = np.random.normal(loc=50, scale=10, size=40)

mean, lower, upper = confidence_interval(sample_data, confidence=0.95)
print(f"Sample mean = {mean:.2f}")
print(f"95% confidence interval = ({lower:.2f}, {upper:.2f})")
```

**10.Write a Python program to calculate the margin of error for a given confidence level using sample data.**


```python
import numpy as np
from scipy.stats import t

def margin_of_error(sample, confidence=0.95):

    n = len(sample)
    std_err = np.std(sample, ddof=1) / np.sqrt(n)  # standard error of mean
    t_crit = t.ppf((1 + confidence) / 2, df=n-1)  # t critical value

    moe = t_crit * std_err
    return moe


np.random.seed(42)
sample_data = np.random.normal(loc=100, scale=15, size=50)

moe = margin_of_error(sample_data, confidence=0.95)
print(f"Margin of error at 95% confidence level: {moe:.3f}")
```

**11.Implement a Bayesian inference method using Bayes' Theorem in Python and explain the process.**


```python
def bayes_theorem(prior, likelihood, false_positive_rate):

    p_data = likelihood * prior + false_positive_rate * (1 - prior)


    posterior = (likelihood * prior) / p_data

    return posterior


prior = 0.01
likelihood = 0.99
false_positive_rate = 0.05

posterior = bayes_theorem(prior, likelihood, false_positive_rate)

print(f"Prior probability (disease prevalence): {prior:.3f}")
print(f"Likelihood (test sensitivity): {likelihood:.3f}")
print(f"False positive rate: {false_positive_rate:.3f}")
print(f"Posterior probability (disease given positive test): {posterior:.3f}")
```

**13.Write a Python program to calculate the expected frequencies for a Chi-square test based on observed data.**


```python
import numpy as np

def expected_frequencies(observed):

    observed = np.array(observed)
    row_sums = observed.sum(axis=1).reshape(-1, 1)
    col_sums = observed.sum(axis=0).reshape(1, -1)
    total = observed.sum()

    expected = (row_sums @ col_sums) / total
    return expected


observed_data = [
    [30, 10, 20],
    [20, 40, 25]
]

expected = expected_frequencies(observed_data)
print("Observed frequencies:\n", np.array(observed_data))
print("Expected frequencies:\n", expected)
```

**14.Perform a goodness-of-fit test using Python to compare the observed data to an expected distribution.**

```python
import numpy as np
from scipy.stats import chisquare


observed = np.array([50, 30, 20])

expected_probs = np.array([0.4, 0.4, 0.2])

expected_counts = expected_probs * observed.sum()

chi2_stat, p_value = chisquare(f_obs=observed, f_exp=expected_counts)

print(f"Observed counts: {observed}")
print(f"Expected counts: {expected_counts}")
print(f"Chi-square Statistic: {chi2_stat:.4f}")
print(f"P-value: {p_value:.4f}")


alpha = 0.05
if p_value < alpha:
    print("Reject null hypothesis: observed data does NOT fit the expected distribution.")
else:
    print("Fail to reject null hypothesis: observed data fits the expected distribution.")
```
